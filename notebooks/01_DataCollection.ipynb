## Download the dataset

The dataset comes from the **Machine Learning Group of ULB** (Université Libre de Bruxelles). It contains transactions made by European cardholders in September 2013.

Due to confidentiality, the original features have been transformed using PCA — so we get V1 through V28 as principal components, plus `Time` and `Amount` in their original form.

```python
# Download and extract the dataset
! kaggle datasets download -d mlg-ulb/creditcardfraud -p data/

# Unzip it
! unzip -o data/creditcardfraud.zip -d data/

# Clean up the zip file
! rm data/creditcardfraud.zip

print("Dataset downloaded successfully.")
```

## Load and inspect the data

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/creditcard.csv")

print(f"Dataset shape: {df.shape}")
print(f"Total transactions: {df.shape[0]:,}")
print(f"Number of features: {df.shape[1]}")
```

```python
# Statistical summary
df.describe()
```

```python
## Class distribution

This is a key thing to check upfront — fraud detection datasets are almost always heavily imbalanced. If there are very few fraud cases relative to legitimate ones, we will need to use techniques like SMOTE during modelling to handle this.
```

```python
fraud_count = df['Class'].sum()
legit_count = (df['Class'] == 0).sum()
fraud_pct = df['Class'].mean()

print(f"Legitimate transactions: {legit_count:,} ({1 - fraud_pct:.2%})")
print(f"Fraudulent transactions: {fraud_count:,} ({fraud_pct:.2%})")
print(f"Imbalance ratio: {legit_count // fraud_count}:1")
```
